### Create Album Summary

```python
from sturdy_barnacle.db_utils import DatabaseManager
from sturdy_barnacle.auto_album_manager import AutoAlbumManager
from sqlalchemy import text

def test_auto_album_summary():

    db = DatabaseManager()
    album_manager = AutoAlbumManager(db)
    
    session = db._get_session()
    album_ids = session.execute(text("SELECT id FROM image_albums")).fetchall()
    session.close()
    
    if not album_ids:
        print("No albums found in the database.")
        return
    
    for album_id in album_ids:
        album_id = album_id[0]
        detected_objects, descriptions, exif_locations = album_manager.get_cluster_metadata(album_id)
        summary = f"This album contains images with objects such as {', '.join(list(detected_objects.keys())[:5])}. "
        
        print(f"Album ID: {album_id}")
        print(f"Generated Summary: {summary}\n")
```

```python
test_auto_album_summary()
```
