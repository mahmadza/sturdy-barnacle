### Query Similar Images
Given an image, query for similar images.

```python
import sys
import os

project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))

sys.path.insert(0, project_root)

from sturdy_barnacle.db_utils import DatabaseManager
from sturdy_barnacle.image_processor import ImageProcessor
from sturdy_barnacle.image_visualizer import ImageVisualizer
```

```python
db = DatabaseManager()
visualizer = ImageVisualizer(db)
```

```python
# Query for images that is similar to image in image_path:
image_path = "/Users/mamduhzabidi/Pictures/Photos Library.photoslibrary/originals/0/02842C50-5BE6-42DF-9E28-F5630910865C.jpeg"
processor = ImageProcessor(image_path, db)
```

```python
embedding = processor.compute_embedding()
results = db.find_similar_images(image_embedding=embedding, top_n=5)
```

```python
# Display similar images
for image in results:
    print(image)
    visualizer.visualize_image_metadata(image['image_path'])
```
