### Query Images Using Keyword
Query images in the database using keywords.

```python
import sys
import os

project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))

sys.path.insert(0, project_root)

from sturdy_barnacle.db_utils import DatabaseManager
from sturdy_barnacle.image_visualizer import ImageVisualizer
```

```python
db = DatabaseManager()
visualizer = ImageVisualizer(db)
```

```python
# Query for images that has keyword "flower"
query_results = db.query_images_by_keyword("flower")
```

```python
for i in range(len(query_results))[:5]:
    image_path = query_results[i].image_path
    visualizer.visualize_image_metadata(image_path)
```
